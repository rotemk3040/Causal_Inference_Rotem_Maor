# Imports

```python
import statsmodels.api as sma
import statsmodels.formula.api as smf
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from tqdm import tqdm
```

# Estimation functions

```python
def estimate_propensity(formula: str, data: pd.DataFrame, show_summary: bool = False) -> pd.Series:
    """
    This function is adopted from Tutorial 4, it estimates a propensity model using generalized linear models. 
    :param formula: string that contains the model formula (in R style) 
    :param data: DataFrame containing the covariates for each observation
    :param show_summary: boolean that indicates to print out the model summary to the console
    :return: propensity scores
    """
    model = smf.glm(formula=formula, data=data, family=sma.families.Binomial()).fit()
    if show_summary:
        print(model.summary())
    return pd.Series(data=model.predict(data), index=data.index)
```

```python
def estimate_with_ipw(data: pd.DataFrame, propensity: pd.Series) -> float:
    """
    This function estimates the ATE and ATT using Inverse Probability Weighting (IPW) with propensity scores.
    The formula is adopted from Tutorial 8.
    :param data: DataFrame containing treatment ("T") and outcome ("Y")
    :param propensity: propensity estimations for all samples
    :return: estimated ATE and ATT
    """
    ate1_comp = sum([t * y / p for p, t, y in zip(propensity, data["T"], data["Y"])]) / len(propensity)
    ate2_comp = sum([((1 - t) * y) / (1 - p) for p, t, y in zip(propensity, data["T"], data["Y"])]) / len(propensity)
    ate_output = ate1_comp - ate2_comp

    att1_comp = np.dot(data["T"], data["Y"]) / np.sum(data["T"])
    att2_comp = np.dot(np.multiply(1 - data["T"], data["Y"]), np.multiply(propensity, 1 / (1 - propensity)))
    att3_comp = np.dot(1 - data["T"], np.multiply(propensity, 1 / (1 - propensity)))
    att_output = att1_comp - (att2_comp / att3_comp)
    return ate_output, att_output.item()
```

```python
#S-learner
def preprocess_slearner(data: pd.DataFrame) -> tuple:
    """
    The function do the preprocessing for the s-learner data
    :param data: DataFrame containing the covariates for each observation
    :return: tuple that contains the treatment data and the Y values 
    """
    new_data = data.copy()

    Y = new_data["Y"]     # save the Y values from the data
    new_data = new_data.drop("Y", axis=1) # remove Y from the data
    return new_data, Y

def fit_model1(data: pd.DataFrame, Y: pd.Series, max_depth: int):
    """
    The function performs decision tree clasification on the data (for the s-learner and t-learner)
    :param data: data contains treatment
    :param Y: data contains the Y values
    :param max_depth: the maximum depth for the random forest regression
    :return: Decision Tree Classifier after the training
    """
    # create the Multi-layer Perceptron regression model and do training
    model = DecisionTreeClassifier(max_depth=3*max_depth, random_state=1)
    model.fit(data.to_numpy(), Y.astype(int).to_numpy())
    return model

def fit_model2(data: pd.DataFrame, Y: pd.Series, max_depth: int):
    """
    The function performs random forest clasification on the data (for the s-learner and t-learner)
    :param data: data contains treatment
    :param Y: data contains the Y values
    :param max_depth: the maximum depth for the random forest regression
    :return: Random forest Classifier after the training
    """
    # create the Multi-layer Perceptron regression model and do training
    model = RandomForestClassifier(max_depth=max_depth, random_state=1)
    model.fit(data.values, Y.astype(int).to_numpy())
    return model
  

def estimate_with_slearner(data: pd.DataFrame, max_depth: int) -> float:
    """
    Estimates the ATE and ATT using S-learner. Recall Toturial 8 for details.
    :param data: DataFrame containing the covariates for each observation
    :param max_depth: the maximum depth for the random forest regression
    :return: estimated ATE and ATT with Decision Tree Classifier and with Random forest Classifier
    """
    # get the data for the s-learner
    s_learner_data, Y = preprocess_slearner(data)

    # get the s-learner model
    model1 = fit_model1(s_learner_data, Y, max_depth)
    model2 = fit_model2(s_learner_data, Y, max_depth)

    # get all the treated data (for ATT we predict on the treated group)
    treated_data = s_learner_data[s_learner_data["T"] == 1]
    # copy the dataset and set the values to zero
    counterfactual_data = s_learner_data.copy()
    counterfactual_data["T"] = 0
    counterfactual_treated_data = treated_data.copy()
    counterfactual_treated_data["T"] = 0

    # compute the ATE, recall Tutorial 8
    predictionATE1 = []
    predictionATE2 = []
    for i in range(s_learner_data.shape[0]):
        val = model1.predict([s_learner_data.iloc[i]]) - model1.predict([counterfactual_data.iloc[i]]) # f(x_i, 1) - f(x_i, 0)
        predictionATE1.append(val)
        val = model2.predict([s_learner_data.iloc[i]]) - model2.predict([counterfactual_data.iloc[i]]) # f(x_i, 1) - f(x_i, 0)
        predictionATE2.append(val)

    # compute the ATT, recall Tutorial 8
    predictionATT1 = []
    predictionATT2 = []
    for i in range(treated_data.shape[0]):
        val = model1.predict([treated_data.iloc[i]]) - model1.predict([counterfactual_treated_data.iloc[i]]) # f(x_i, 1) - f(x_i, 0)
        predictionATT1.append(val)
        val = model2.predict([treated_data.iloc[i]]) - model2.predict([counterfactual_treated_data.iloc[i]]) # f(x_i, 1) - f(x_i, 0)
        predictionATT2.append(val)

    return np.mean(predictionATE1), np.mean(predictionATT1), np.mean(predictionATE2), np.mean(predictionATT2)
```

```python
def preprocess_tlearner(data: pd.DataFrame) -> tuple:
    """
    The function do the preprocessing for the t-learner data
    :param data: DataFrame containing the covariates for each observation
    :return: tuple that contains the treatment data and the Y values for the treated data (T == 1) and 
    control data (T == 0) 
    """
    # split the data
    treated_data = data[data["T"] == 1]
    control_data = data[data["T"] == 0]
    treated_data, treated_Y = preprocess_slearner(treated_data)
    control_data, control_Y = preprocess_slearner(control_data)
    return treated_data, treated_Y, control_data, control_Y


def estimate_with_tlearner(data: pd.DataFrame, max_depth: int) -> float:
    """
    Estimates the ATE and ATT using T-learner. Like the S-learner, recall Tutorial 8 for details.
    :param data: data contains treatment and outcome
    :param max_depth: the maximum depth for the random forest regression
    :return: estimated ATE and ATT with Decision Tree Classifier and with Random forest Classifier
    """
    # get the data
    treated_data, treated_Y, control_data, control_Y = preprocess_tlearner(data)
    whole_data, whole_y = preprocess_slearner(data)

    # get a regression model for the treated and control data
    model1_treated = fit_model1(treated_data, treated_Y, max_depth)
    model1_control = fit_model1(control_data, control_Y, max_depth)
    model2_treated = fit_model2(treated_data, treated_Y, max_depth)
    model2_control = fit_model2(control_data, control_Y, max_depth)


    # compute the ATT
    predictions1, predictions2 = [], []
    for i in range(treated_data.shape[0]):
        val = model1_treated.predict([treated_data.iloc[i]]) - model1_control.predict([treated_data.iloc[i]]) #f_1(x_i) - f_0(x_i)
        predictions1.append(val)
        val = model2_treated.predict([treated_data.iloc[i]]) - model2_control.predict([treated_data.iloc[i]]) #f_1(x_i) - f_0(x_i)
        predictions2.append(val)

    ATT1, ATT2 = np.mean(predictions1), np.mean(predictions2)
    #compute the ATE

    for i in range(control_data.shape[0]):
        val = model1_treated.predict([control_data.iloc[i]]) - model1_control.predict([control_data.iloc[i]]) #f_1(x_i) - f_0(x_i)
        predictions1.append(val)
        val = model2_treated.predict([control_data.iloc[i]]) - model2_control.predict([control_data.iloc[i]]) #f_1(x_i) - f_0(x_i)
        predictions2.append(val)
    
    ATE1, ATE2 = np.mean(predictions1), np.mean(predictions2)
    

    return ATE1, ATT1, ATE2, ATT2
```

```python
def matching(label: pd.Series, propensity: pd.Series, calipher: float = 0.0):
    """
    Performs nearest-neighbour matching for a sample of test and control observations,
    based on the propensity scores for each observation. Adopted from tutorial 4.
    :param label: series that contains the label for each observation
    :param propensity: series that contains the propensity score for each observation
    :param calipher: bound on distance between observations in terms of propensity score
    :param replace: boolean that indicates whether sampling is with (True) or without replacement (False)
    :return: matches
    """
    treated = propensity[label == 1]
    control = propensity[label == 0]

    # Randomly permute in case of sampling without replacement to remove any bias arising from the ordering of the data set
    matching_order_treated = np.random.permutation(label[label == 1].index)
    matching_order_control = np.random.permutation(label[label == 0].index)
    matches_treated, matches_control = {}, {}

    for obs in matching_order_treated:
        # Compute the distance between the treatment observation and all candidate controls in terms of propensity score
        distance = abs(treated[obs] - control)

        # Take the closest match
        if distance.min() <= calipher or not calipher:
            matches_treated[obs] = [distance.idxmin()]
    
    for obs in matching_order_control:
        # Compute the distance between the treatment observation and all candidate controls in terms of propensity score
        distance = abs(control[obs] - treated)

        # Take the closest match
        if distance.min() <= calipher or not calipher:
            matches_control[obs] = [distance.idxmin()]

    return matches_treated, matches_control


def estimate_with_matching(data: pd.DataFrame, propensity: pd.Series) -> float:
    """
    Estimates the ATE and ATT using matching.
    :param data: data contains treatment and outcome
    :param propensity: propensity estimations for all samples
    :return: estimated ATE and ATT
    """
    matches_treated, matches_control = matching(data["T"], propensity)
    att = np.mean([data["Y"][key] - data["Y"][val[0]] for key, val in matches_treated.items()]) # key= treatment observation, val[0] = closest control
    ate = np.mean([data["Y"][key] - data["Y"][val[0]] for key, val in matches_treated.items()] + [data["Y"][key] - data["Y"][val[0]] for key, val in matches_control.items()])
    return ate.item(), att.item()
```

```python
#constants
NUM_COVARIATES = len(filtered_df.columns) - 2
att_results_path = "ATT_results.csv"
propensity_results_path = "models_propensity.csv"
max_depth = 10

# lists to save the results
propensity_results = []
att_results = []

# copy the filtered data frame and preprocess
data = filtered_df.copy()
data = data.rename(columns = {'Target' : 'Y', 'Displaced' : 'T'})
data['Y'][data['Y'] != 'Graduate'] = 0
data['Y'][data['Y'] == 'Graduate'] = 1
COLS = list(data.columns)
for x in ['T','Y']: COLS.remove(x) 
COLS += ['T','Y']
data=data.reindex(columns = COLS)
data = data.rename(columns = {data.columns[i] : 'x_'+str(i) for i in range(len(filtered_df.columns) - 2)})

# calculate propensity scores
print("Estimating propensities")
formula = f"T ~ {' + '.join(data.columns[0:NUM_COVARIATES])}"
propensity_results = estimate_propensity(formula, data)

# 1 - estimate ATEs ATTs with IPW
print("Estimating IPW")
att_results.append(estimate_with_ipw(data, propensity_results))

# 2 - estimate ATEs and ATTs with S-learner
print("Estimating S-Learner")
att_results.append(estimate_with_slearner(data, max_depth = max_depth))

# 3 - estimate ATEs and ATTs with T-learner
print("Estimating T-Learner")
att_results.append(estimate_with_tlearner(data, max_depth=max_depth))

# 4 - estimate ATEs and ATTs with matching
print("Estimating matching")
att_results.append(estimate_with_matching(data, propensity_results))
```
